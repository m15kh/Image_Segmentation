```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_paths = '../input/PennFudanPed/train_images'
mask_paths = '../input/PennFudanPed/train_masks'

all_images = os.listdir(image_paths)
all_masks = os.listdir(mask_paths)

all_images.sort()
all_masks.sort()
```

```python
for i, data in enumerate(all_images):
    image = cv2.imread(os.path.join(image_paths, all_images[i]))
    mask = cv2.imread(os.path.join(mask_paths, all_masks[i]))
    # Make all pixel > 0 as 255.
    im = mask > 0
    mask[im] = 255
    mask[np.logical_not(im)] = 0

    plt.figure(figsize=(10, 7))
    for j in range(2):
        plt.subplot(1, 2, 1)
        plt.imshow(image[..., ::-1])
        plt.axis('off')
        plt.subplot(1, 2, 2)
        plt.imshow(mask)
        plt.axis('off')
    plt.show()
    if i == 3:
        break
```
